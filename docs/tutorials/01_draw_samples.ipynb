# Draw completed BOSS CMASS-South catalogs

Every ECHOES realization is an equal-weight, cosmology-free completed catalog.
We only need the 2 MB posterior package and NumPy.

```python
from echoes import load_package, draw
import numpy as np
pkg = load_package('../../data_release/cmass_south_posterior.npz')
cat = draw(pkg, seed=0)
print(cat['N'], 'galaxies')
for code, name in zip(*np.unique(cat['prov'], return_counts=True)):
    print(int(code), int(name))
```

`PROV`: 0 observed · 1 fiber-collided · 2 redshift-failure · 3 imaging-systematic analog.

A reproducible ensemble is just a list of seeds; the observed galaxies are shared,
only the missing redshifts vary. Run *your* statistic across seeds and use the
scatter as the completion uncertainty.

```python
ensemble = [draw(pkg, seed=s) for s in range(20)]
nz = np.mean([np.histogram(c['z'], bins=30, range=(0.43,0.7))[0] for c in ensemble], axis=0)
print('mean n(z):', nz.astype(int))
```
